```python
nums = [2, 5, 8, 3] 
#nums adlı içində müxtəlif rəqəmlər olan list yaratdıq
nums[1] = nums[1] * 2
#bu listin birinci indexde olan elementini 2-ye vurduq yeni 5*2 = 10
nums.append(10) 
#liste sonuncu element olaraq 10 elementini elave etdik
nums.insert(0, 0) 
#listin sifirinci indexine 0 elementini elave etdik, netice olaraq butun elementler bir vahid saga surusmus oldu
print(nums) 
#nums listini cap eledik netice olaraq [0, 2, 10 burada olan 10 5*2-den alinib, 8, 3, 10 append edende gelib en sona dusur]
```

```python
text = "python is amazing" 
#texte pytho is amazing adli string yaradib bu deyeri assing  eledik
result = "" 
#result deyiseni ucun bos string yaratdiq
for ch in text: 
#texti yoxlamaq ucun loop yaratdiq
if ch != " ": 
#icinde olan bosluqlari silir ifdeki " " bu serte beraber deyildire gore
result += ch.upper() 
#netice olaraq result deyisenine gelir boyuk herflerle yazilmis PYTHONISAMAZING stringini assing edir
print(result)
#burada ise qeyd etdiyimiz kimi resultu bize cap edir
#output → PYTHONISAMAZING 
```

```python
nums = [1, 4, 7, 12, 15, 18] 
#geldik nums adli list yaratdiq
filtered = [] 
#filtered adli bos list yaratdiq
for n in nums: 
#burada for loopu yaradiriq hansi ki n gelib numsin icindeki ededleri bir bir goturur
if n > 10:
#burada sert olaraq veririk ki, eger n 10-dan boyukdurse
filtered.append(n * 2)
#n 10-dan boyukdurse filtered listine ikiye vuraraq elave etmeyimizi isteyir
print(filtered) 
#output → [24, 30, 36]
```

```python
nums = [3, 1, 4, 1, 5, 9] 
# nums adli list yaratdiq
result = 0 
#resulta biz sifir verdik
for n in nums: 
#for loopu yaratdiq n gelib bir-bir nums listindeki deyerleri cekecek
if n % 2 != 0: 
#sert qoyduq eger n-i ikiye bolende qaliq sifira beraber deyilse
result += n - 1 
#resultun uzerinde herdefesinde n-1-i elave edib ona beraber edesiyik
print(result) 
#resultu capa veriririk netice olaraq 14 alinir cunki 3-1, 1-1, 1-,1 5-1, 9-1 -i toplayiriq bu zaman 14 alinir cavab
output → 14 
```
