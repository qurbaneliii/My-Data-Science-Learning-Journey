```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle, Arc

# Pitch dimensions (meters)
L, W = 105, 68

fig, ax = plt.subplots(figsize=(12, 8))

# Background (grass)
ax.add_patch(Rectangle((0, 0), L, W, facecolor="#2e7d32", edgecolor="none"))

# Outer lines
ax.add_patch(Rectangle((0, 0), L, W, fill=False, edgecolor="white", linewidth=2))

# Halfway line
ax.plot([L/2, L/2], [0, W], color="white", linewidth=2)

# Center circle + center spot
ax.add_patch(Circle((L/2, W/2), 9.15, fill=False, edgecolor="white", linewidth=2))
ax.add_patch(Circle((L/2, W/2), 0.25, color="white"))

# Penalty areas (16.5m) and goal areas (5.5m)
# Left side
ax.add_patch(Rectangle((0, (W-40.32)/2), 16.5, 40.32, fill=False, edgecolor="white", linewidth=2))
ax.add_patch(Rectangle((0, (W-18.32)/2), 5.5, 18.32, fill=False, edgecolor="white", linewidth=2))
ax.add_patch(Circle((11, W/2), 0.25, color="white"))  # penalty spot

# Right side
ax.add_patch(Rectangle((L-16.5, (W-40.32)/2), 16.5, 40.32, fill=False, edgecolor="white", linewidth=2))
ax.add_patch(Rectangle((L-5.5, (W-18.32)/2), 5.5, 18.32, fill=False, edgecolor="white", linewidth=2))
ax.add_patch(Circle((L-11, W/2), 0.25, color="white"))  # penalty spot

# Penalty arcs (radius 9.15m) outside the box
ax.add_patch(Arc((11, W/2), 18.3, 18.3, angle=0, theta1=310, theta2=50, color="white", linewidth=2))
ax.add_patch(Arc((L-11, W/2), 18.3, 18.3, angle=0, theta1=130, theta2=230, color="white", linewidth=2))

# Goals (approx visual, depth 2m, width 7.32m)
goal_w = 7.32
goal_d = 2.0
ax.add_patch(Rectangle((-goal_d, (W-goal_w)/2), goal_d, goal_w, fill=False, edgecolor="white", linewidth=2))
ax.add_patch(Rectangle((L, (W-goal_w)/2), goal_d, goal_w, fill=False, edgecolor="white", linewidth=2))

# Corner arcs (radius 1m)
r = 1.0
ax.add_patch(Arc((0, 0), 2*r, 2*r, theta1=0, theta2=90, color="white", linewidth=2))
ax.add_patch(Arc((0, W), 2*r, 2*r, theta1=270, theta2=360, color="white", linewidth=2))
ax.add_patch(Arc((L, 0), 2*r, 2*r, theta1=90, theta2=180, color="white", linewidth=2))
ax.add_patch(Arc((L, W), 2*r, 2*r, theta1=180, theta2=270, color="white", linewidth=2))

# Optional: mowing stripes
for i in range(0, L, 10):
    if (i // 10) % 2 == 0:
        ax.add_patch(Rectangle((i, 0), 10, W, facecolor="#2f8f37", edgecolor="none", alpha=0.35))

# Formatting
ax.set_xlim(-3, L+3)
ax.set_ylim(-3, W+3)
ax.set_aspect('equal', adjustable='box')
ax.axis("off")

plt.show()
```
