```python
import sys 
import numpy as np

from PyQt6.QtWidgets import QApplication
from pyqtgraph.Qt import QtCore
import pyqtgraph.opengl as gl

def lorenz_step(x, y, z, sigma= 10.0, beta = 8.0/3.0, rho = 28.0):
    dx = sigma * (y - x)
    dy = x*(rho-z)-y
    dz=x*y-beta*z
    return dx, dy, dz

def main():
    app = QApplication(sys.argv)
    w = gl.GLViewWidget()
    w.show()
    w.setWindowTitle('Lorenz Attractor')
    w.setCameraPosition(distance=80, elevation=18, azimuth=35)
    w.setBackgroundColor('k')
    
    axis = gl.GLAxisItem()
    axis.setSize(30, 30, 30)
    w.addItem(axis)
    
    sigma, beta, rho = 10.0, 8.0/3.0, 28.0
    dt = 0.01
    num_particles = 1500
    
    x = np.random.uniform(-10, 10, num_particles)
    y = np.random.uniform(-10, 10, num_particles)
    z = np.random.uniform(-10, 10, num_particles)
    
    # Define pos array
    pos = np.column_stack((x, y, z)).astype(np.float32)
    
    colors = np.ones((num_particles, 4), dtype=np.float32)
    colors[:, :3] = 0.9
    colors[:, 3] = 0.9
    
    scatter = gl.GLScatterPlotItem(pos=pos, color=colors, size=2.0, pxMode=True)
    w.addItem(scatter)
    
    fade = 0.985
    
    def update():
        nonlocal x, y, z, pos, colors
        for i in range(num_particles):
            dx, dy, dz = lorenz_step(x[i], y[i], z[i], sigma, beta, rho)
            x[i] += dx * dt
            y[i] += dy * dt
            z[i] += dz * dt
        
        pos[:, 0] = x
        pos[:, 1] = y
        pos[:, 2] = z

        z_norm = (z - z.min()) / (z.max() - z.min() + 1e-9)
        colors[:, 0] = 0.2 + 0.8 * z_norm
        colors[:, 1] = 0.3 + 0.6 * (1 - z_norm)
        colors[:, 2] = 0.9
        colors[:, 3] *= fade
        low = colors[:, 3] < 0.15
        colors[low, 3] = 0.9
        scatter.setData(pos=pos, color=colors)
    
    timer = QtCore.QTimer()
    timer.timeout.connect(update)
    timer.start(16)
    sys.exit(app.exec())

if __name__ == '__main__':
    main()
```
